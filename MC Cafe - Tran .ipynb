```python
#coffee dictionary
coffee_menu = {
    "espresso": 2.50,
    "cappuccino": 3.00,
    "latte": 3.50,
    "mocha": 4.00,
    "americano": 2.75
}

#tea dictionary
tea_menu = {
    "white tea": 1.50,
    "green tea": 2.00,
    "oolong tea": 2.25,
    "black tea": 2.25,
    "earl grey": 2.50,
    "matcha": 2.50
}

#pastry dictionary
pastry_menu = {
    "croissant": 5.50,
    "cinnamon roll": 4.50,
    "banana bread": 4.25,
    "pumpkin bread": 3.75,
    "sticky bun": 4.25,
    "apple turnover": 3.75
}



#function to calculate the order total passing the 3 menus as parameters
def calculate_total(order, coffee_menu, tea_menu, pastry_menu): #aka can also pass as '*menus' or 'menu1,menu2,menu3'
    subtotal = 0
    order_receipt = {} #empty dictionary to store ordered items and their quantity 
    for menu in [coffee_menu, tea_menu, pastry_menu]: #outer loop that iterates over each of the menus
        for item, quantity in order.items(): #inner loop that iterates over the items in the customer's order
            if item in menu: #if item found in menu, it calculates subtotal and updates order_receipt
                subtotal += menu[item] * quantity
                order_receipt[item] = quantity  
    return subtotal, order_receipt #enables recall of the total cost, items ordered, and their quantities.


#main function that displays menu and takes customer order
def main():
    print("Welcome to our coffee shop!")
    print("Here is our menu:")
    print("\n-- Coffee Menu --")
    for item, price in coffee_menu.items():
        print(f"{item}: ${price:.2f}")
    print("\n-- Tea Menu --")
    for item, price in tea_menu.items():
        print(f"{item}: ${price:.2f}")
    print("\n-- Pastry Menu --")
    for item, price in pastry_menu.items():
        print(f"{item}: ${price:.2f}")
        
    customer_order = {}
   
    while True:
        item = input("\nWhat would you like to order? (Type 'done' to finish): ").strip().lower()
        if item == 'done':
            break
        if item in coffee_menu or item in tea_menu or item in pastry_menu:
            quantity = int(input(f"How many {item}s would you like? "))
            if item in customer_order:
                customer_order[item] += quantity
            else:
                customer_order[item] = quantity
        else:
            print("Sorry, that item is not on the menu. Please choose an item from the menu.")
            
      
       

    #math portion
    subtotal, order_receipt = calculate_total(customer_order, coffee_menu, tea_menu, pastry_menu)   
     
    print("\n--- Order Receipt ---")
    for item, quantity in order_receipt.items():
        print(f"{item}: {quantity}")
    print(f"\nSubtotal: ${subtotal:.2f}")
    
    #calculate sales tax
    sales_tax_rate = 0.06  # Maryland sales tax rate
    sales_tax = subtotal * sales_tax_rate
    print(f"Sales Tax (6%): ${sales_tax:.2f}")

    #calculate grand total
    grand_total = subtotal + sales_tax
    print(f"Grand Total: ${grand_total:.2f}")
    
    print("\nThank you for visiting our coffee shop!")
    

main()
```
